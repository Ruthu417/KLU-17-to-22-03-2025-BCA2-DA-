#### Introduction to Data Analysis
- Data is the information stored in a file 
- The complete study of data is called data science
- Machine Learning and Artificial Intelligence are the parts of data science

- Biodata / personal info
    - Ex:Name,gender,mobile ,identity(aadhar),contact address,designation/occupation etc
- Data in realtime
    1. Discrete
        - this type of data is fixed/constant/static
        - Name,aadhar etc.
        - 2types of data
            1. Quantitative
                - the data is organized on the basis of quantity-->size/volume
                - file size,weight etc
            2. Qualitative
                - the data is categorized based on the quality
                - feedback 
            
    2. Continuous
        - age,height,weight,time etc.
        
- Data in our devices
    - .jpg,.jpeg,.png(image)
    - .mp3,.mp4,.txt,.doc,.xls,.exe,.py,.c,
    - .apk file
- Data in statistics
    - 4types of data
    1. Nominal
        - data can be organized based on the names
    2. Ordinal
        - order of the data points
            - ascending
            - descending
    3. Rational
        - p/q (ratio/proportion) of data
    4. Interval
        - data present in a range
